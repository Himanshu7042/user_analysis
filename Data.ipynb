```python
import pandas as pd
file = pd.read_excel("/content/Active Users (4) (1) (1) (1) (1).xls")
```

```python
file.head()
```

```python
file = file.drop(columns=['Unnamed: 56','Unnamed: 57','Unnamed: 58',	'Unnamed: 59'])
file.head()
```

```python
file['w1'].shape
```

```python
col = file.columns
```

```python
non_null_counts = file.notnull().sum()
non_null_counts.head()
```

```python
new_df = pd.DataFrame({
    'Week': col,
    'MAU': non_null_counts
})
new_df.head()
```

```python
new_df = new_df.reset_index(drop=True)
new_df.head()
```

```python
new_df['New'] = 0
new_df['Retained'] = 0
new_df['Resurrected'] = 0
new_df['Churned'] = 0

new_df['Retention Rate'] = 0
new_df['Quick Ratio'] = 10

previous_week_users = set()
old_week_users = set()
previous_MAU = 0

for i, week in enumerate(new_df['Week']):

  current_week_users = set(file[week].dropna())     #dropping null values

  new_users_count = 0
  retained_users_count = 0
  resurected_users_count = 0
  if week=='w1':
    new_users_count = len(current_week_users)

  else:
    new_users_count = len(current_week_users - previous_week_users - old_week_users)
    retained_users_count = len(current_week_users.intersection(previous_week_users))
    resurected_users_count = new_df['MAU'] - new_users_count - retained_users_count

    #Churned
    churned_users_count = new_users_count + resurected_users_count + previous_MAU - new_df['MAU']
    new_df.loc[new_df['Week'] == week, 'Churned'] = churned_users_count

    #Retention Rate
    retention_rate = (retained_users_count / len(previous_week_users)) * 100 if len(previous_week_users) > 0 else 0
    new_df.loc[new_df['Week'] == week, 'Retention Rate'] = retention_rate

    #Quick Ratio   (new+resurected)/churned
    new_df.loc[new_df['Week'] == week, 'Quick Ratio'] = (new_users_count+resurected_users_count)/churned_users_count

  new_df.loc[new_df['Week'] == week, 'New'] = new_users_count
  new_df.loc[new_df['Week'] == week, 'Retained'] = retained_users_count
  new_df.loc[new_df['Week'] == week, 'Resurrected'] = resurected_users_count


  #Updating perameters for next itertion
  previous_week_users = current_week_users
  old_week_users = old_week_users.union(previous_week_users)
  previous_MAU = len(current_week_users)
```

```python
new_df.head()
```

```python
new_df.to_csv("/content/users_data.csv",index=False)
```
